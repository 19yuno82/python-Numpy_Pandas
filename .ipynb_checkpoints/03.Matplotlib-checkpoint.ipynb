### matplotlib
- 2D 그래프로 시각화가 가능한 라이브러리
- 파이썬에서 데이터를 차트로 그려줌
- 차트의 종류는 산점도, 막대, 파이, 히스토그램, 꺽은선 그래프 등 다양함

### 데이터 시각화란?
- 광범위하게 분산된 방대한 양의 자료를 한눈에 볼 수 있도록 도표나 차트 등으로 정리하는 것
- 시각화를 통해 데이터의 특징을 쉽게 파악할 수 있다.

```python
import matplotlib.pyplot as plt
```

#### 기본적인 그래프 그려보기

```python
# Line Plot (선 그래프)
y = [2,12,6]
plt.plot(y)
#X좌표는 0부터 1씩 증가하며 자동 할당
plt.show()
#그렸던 plot(그래프)들을 출력
```

```python
#x축, Y축 설정해서 그려보기
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,ls=':',color='r',lw=2, marker='D',mec='b',mfc='y')
# 속성값
# 1. 라인의 스타일 설정 : ls(Line Style) -> 문자열
# 2. 라인의 두께 지정 : lw(Line Width) -> 그냥 숫자로 적어주기!! 문자X
# 3. 선의 색 바꾸기 : c(color) -> 선색과 마커의 색까지 모두 변경
# 4. 값의 위치 표시 : marker -> 문자열
# 5. 마커 내부 색깔 : mfc(marker face color) -> 문자열
# 6. 마커 선 색깔 : mec(marker edge color) -> 문자열
# 7. 마커 크기 : ms -> 숫자
# 8. 마커 선 굵기 : mew(marker egde width) -> 숫자
plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]
#색상코드표를 활용해서 색상 지정 가능
#https://www.rapidtables.org/ko/web/css/css-color.html
plt.plot(x,y,ls='-.',color='m',marker='*',mec='r',ms=15)
#그래프 숫자 표시 범위 지정(xlim,ylim)
#범위를 지정하여 그래프 확대, 축소 출력 가능
plt.xlim(-2, 10)
plt.ylim(-5, 12)
#그리드 그리기(모눈종이,격자눈금)
plt.grid()
#눈금 출력하기
plt.yticks(y) #y데이터 표시

plt.show()
```

#### 여러개 차트를 하나의 공간에 그리기

```python
#한글 인코딩하는 법 - matplotlib 한글이 깨졌을 때 사용
from matplotlib import rc
rc('font', family= 'malgun Gothic')
#font : 폰트를 설정하겠다
#family: 글꼴집합이라는 의미로 글꼴명을 적어주면 됨
```

```python
x=[2,3,4,5,6,7,8]
y1=[5,7,3,9,1,8,6]
y2=[9,5,7,2,3,7,1]

plt.plot(x,y1)
plt.plot(x,y2)

#x축 이름 설정
plt.xlabel('x축')
#y축 이름 설정
plt.ylabel('y',rotation=0)
plt.show()
```

```python
#여러 개의 차트 하나의 공간에 그리기
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x,label='x data')
plt.plot(y,label='y data')
plt.plot(z,label='z data')

#범례 설정하기
#범례 : 참고사항, 데이터를 식별하기 위한 text
plt.legend()

#차트 제목 설정하기
plt.title('차트 제목',loc='right')

plt.show()
```

#### 산점도(scatter) 그래프

```python
#DataFrame 생성
import pandas as pd

df = pd.DataFrame(np.random.rand(50,4), columns=['A','B','C','D'])
df.head(5) #최상단에 있는 데이터 확인
df.tail(3) #최하단에 있는 데이터 확인
```

```python
#scatter 그래프 생성
plt.scatter(x=df['A'],y=df['B'],label='Group1',s=10,c='aqua')
plt.scatter(x=df['C'],y=df['D'],label='Group2',s=20)

plt.legend()

plt.show()
```

#### bar 그래프 그려보기

```python
#바(막대) 그래프
x=[0,1,2,3,4,5]
y=[80,95,70,60,50,90]

plt.bar(x,y,color='coral')
plt.xlabel('스인개 반')
plt.ylabel('점수',rotation=0)
#plt.text(x좌표,y좌표,적고자 하는 값)
for i in range(len(x)) :
    plt.text(x[i]-0.1,y[i]-5,f'{y[i]}')
#ylim을 설정해서 막대그래프 확대
plt.ylim(45,100)

#제목 달아주기
plt.title('스인개 반', loc='left')

plt.show()
```

```python
# barh(수평막대) 그래프
x=[0,1,2,3,4,5]
y=[80,95,70,60,50,90]

plt.barh(x,y)

plt.show()
```

### 교통사고 사망자수를 count해보자

### 요일별 사고건수를 카운트
- 요일별 사고건수 확인
- 요일 순서대로(월~일) 출력

```python
y = data['요일'].value_counts()
```

```python
#재정렬 하는 방법
y=y[['월','화','수','목','금','토','일']]
```

#### 시각화
- 막대그래프

```python
plt.bar(y.index,y.values, color='gold')
plt.ylim(450,650)
for i in range(y.size) :
    plt.text(i-0.2,y[i],f'{y[i]}')
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 교통사고 건수')


plt.show()
```

### '차대차'(차와 차의 사고) 사건 중 사상자 수가 많은 발생지 시도를 알아보고 시각화
- 키워드 : 차대차, 사상자수,발생지 시도

```python
data.columns
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
car = data[data['사고유형_대분류'] == '차대차']
```

```python
car[['사상자수','발생지시도']]
```

```python
#빌생지시도를 기준으로 사상자수를 확인
car1 = car[['사상자수','발생지시도']].groupby(['발생지시도']).sum()
```

```python
plt.plot(car1.index,car1.values)
plt.title('차vs차 교통사고의 사상자수')
plt.xlabel('지역')
plt.ylabel('사상자 수',rotation=0)

plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.bar(car1.index,car1['사상자수'])
plt.title('차vs차 교통사고의 사상자수')
plt.xlabel('지역')
plt.ylabel('사상자 수',rotation=0)

plt.show()
```

#### 그래프를 내림차순으로 정렬

```python
car1 = car1.sort_values('사상자수', ascending=False)
```

```python
plt.figure(figsize=(15,8))

plt.bar(car1.index,car1['사상자수'])
plt.title('차vs차 교통사고의 사상자수')
plt.xlabel('지역')
plt.ylabel('사상자 수',rotation=0)

plt.show()
```
